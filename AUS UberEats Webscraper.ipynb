# Functions

```python
def search_box():
    search_box = wait.until(EC.element_to_be_clickable((By.XPATH, f'/html/body/div[1]/div[1]/div/main/div[1]/div[2]/div/div[1]/div[1]/div/div/input')))
    return search_box
```

```python
def search_city(namesame):
    search_city = search_box().send_keys(name)
    return search_city
```

```python
def find_food_button():
    find = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div/main/div[1]/div[2]/div/div[1]/button')))
    return find
```

```python
def show_more():
    more = True
    while more == True:
        try:
            show_more = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div/main/div/div[3]/div[2]/div/button'))).click()
        except:
            driver.find_element_by_tag_name('body').send_keys(Keys.CONTROL + Keys.HOME)
            more = False
```

```python
def card(x):
    card = wait.until(EC.element_to_be_clickable((By.XPATH, f'/html/body/div[1]/div[1]/div/main/div/div[3]/div[2]/div/div[2]/div[{x}]/div/a/h3')))
    return card
```

```python
def restaurant_name():
    name = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[1]/div/main/div[3]/div/div/div[3]/div[1]/div[2]/div[2]/h1'))).text
    return name
```

```python
def cuisine_types():
    try:
        cuisine_list = []
        x = 1
        boolean = True
        while boolean == True:
            try:
                cuisines = wait.until(EC.presence_of_element_located((By.XPATH, f'/html/body/div[1]/div[1]/div/main/div[4]/div/div[1]/div/a[{x}]'))).text
                cuisine_list.append(cuisines)
                x += 1                                                    
            except:
                boolean = False

        cuisines = ','.join(cuisine_list)
        return cuisines
    except:
        return 'NULL'
```

```python
def rating():
    try:
        rating = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[1]/div/main/div[3]/div/div/div[3]/div[1]/div[2]/div[2]/div[1]/div[7]'))).text
        if '.' and not '$' in rating:
            return rating
        else:
            return 'NULL'
    except:
        return 'NULL'
```

```python
def address():
    try:
        address = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[1]/div/main/div[4]/div/div[1]/p'))).text
        address = ' '.join(address.split(' ')[:-1])
        return address
    except:
        return 'NULL'
```

```python
def link():
    link = driver.current_url
    return link
```

```python
def if_closed():
    try:
        closed = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[1]/div/div[5]/div/div/div[2]/div[2]/button')))
        return closed
    except:
        return 'NULL'
```

```python
def previous_page():
    driver.execute_script("window.history.go(-1)")
```

```python
def search_address_sequence(name):
    search_city(name)
    time.sleep(1)
    find_food_button().click()
    show_more()
```

```python
def search_restaurants_sequence():
    x = 1
    y = 1
    condition = 'True'
    boolean = True
    while boolean == True:
        try:
            card(x).click()
            if if_closed() != 'NULL':
                print('we in here')
                previous_page()
            print('=================================1')
            print(restaurant_name())
            print(cuisine_types())
            print(rating())
            print(address())
            print(link())
            areas.append(name)
            restaurant_names.append(restaurant_name())
            cuisines_types.append(cuisine_types())
            ratings.append(rating())
            addresses.append(address())
            links.append(link())
            print('=================================2')
            previous_page()
            x += 1
            y = 0
        except:
            print(f'searching... {x}')
            x += 1
            y += 1
            condition = 'False'
        if y > 25 and condition == 'False':
            boolean = False
    print(f'{x} out of 200 complete')
```

# Main

```python
for name in names[0:10]:
    path = r"C:\Users\Abdulkadir\Documents\Programming\chromedriver.exe"
    driver = webdriver.Chrome(path)
    driver.get('https://www.ubereats.com/au')
    wait = WebDriverWait(driver, 5)
    time.sleep(1)
    search_address_sequence(name)
    search_restaurants_sequence()
    
    driver.close()
```
