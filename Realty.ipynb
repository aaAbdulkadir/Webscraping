```python
# ------------------------------------------------------- IMPORT -------------------------------------------------------
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
from selenium.webdriver.support import expected_conditions as EC, wait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support.ui import WebDriverWait
import requests

# ------------------------------------------------------- GLOBAL VARIABLE ------------------------------------------------
expiry_dates = []

years = []
residentials = []
commercials = []
totals = []

projects = []
transactions = []

rankings = []

names1 = []
names2 = []
names3 = []
# ------------------------------------------------------- MAIN ------------------------------------------------

# Get all the names from the csv to loop through
csv = pd.read_csv(r"C:\Users\Abdulkadir\Documents\Programming\WebScrapeTeasdale\Realty\names_list.csv")
names = csv["Name"].to_list()

if __name__ == "__main__":
    
    for name in names[2000:2050]:
        # open website
        path = r"C:\Users\Abdulkadir\Documents\Programming\chromedriver.exe"
        driver = webdriver.Chrome(path)
        driver.get('https://dubailand.gov.ae/en/eservices/licensed-real-estate-brokers/licensed-real-estate-brokers-list/#/')
        wait = WebDriverWait(driver, 10)
        driver.maximize_window()

        # get rid of pop up
        pop_up = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[18]/div/div[1]/a')))
        time.sleep(1)
        pop_up.click()
        
        # search box
        search_box = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[7]/div[3]/div/div[2]/div[1]/div/label/input')))
        time.sleep(4)
        search_box.click()
        search_box.send_keys(name)

        # click name
        e = 1
        time.sleep(2)
        brokerTrue = True
        while brokerTrue == True:
            # if card there
            try:
                try:
                    broker_card = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="broker"]/div/div/div/div/a')))
                    actions = ActionChains(driver)
                    actions.move_to_element(broker_card).perform()
                    broker_card.click()
                    brokerTrue = False
                except:
                    print('try again mayn')
                    e += 1
            
            # if card not there
            except:
                brokerTrue = False

        print('------------------------------------------------------------')

        # office expiry
        office_tab = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/nav/ul/li[2]/a')))
        office_tab.click()

        expiry_date = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/div[2]/div[2]/div[2]/p[2]/span'))).text
        print(expiry_date)
        expiry_dates.append(expiry_date)

        print('------------------------------------------------------------')

        # sales transactions
        transactions_tab = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/nav/ul/li[3]/a')))
        transactions_tab.click()

        a = 0
        b = 0
        y = 1

        transactionsTrue = True
        while transactionsTrue == True:
            try:
                
                year = wait.until(EC.presence_of_element_located((By.XPATH, f'/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/div[3]/div/table/tbody/tr[{y}]/td[1]'))).text
                print(year)
                years.append(year)
                
                if year == 'No Transactions Found':
                    residentials.append('N/A')
                    commercials.append('N/A')
                    totals.append('N/A')
                    names1.append(name)
                    transactionsTrue = False

                residential = wait.until(EC.presence_of_element_located((By.XPATH, f'/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/div[3]/div/table/tbody/tr[{y}]/td[2]'))).text
                print(residential)
                residentials.append(residential)

                commercial = wait.until(EC.presence_of_element_located((By.XPATH, f'/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/div[3]/div/table/tbody/tr[{y}]/td[3]'))).text
                print(commercial)
                commercials.append(commercial)

                total = wait.until(EC.presence_of_element_located((By.XPATH, f'/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/div[3]/div/table/tbody/tr[{y}]/td[4]'))).text
                print(total)
                totals.append(total)
                
                names1.append(name)

                y += 1

            except:
                year = 'No'
                y += 1

            b += 100
            driver.execute_script(f"window.scrollTo({str(a)},{str(b)})")

            if year == 'No':
                transactionsTrue = False

        print('------------------------------------------------------------')

        # projects
        projects_tab = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/nav/ul/li[4]/a')))
        projects_tab.click()

        a = 0
        b = 0
        y = 1

        projectsTrue = True
        while projectsTrue == True:
            try:    
                project = wait.until(EC.presence_of_element_located((By.XPATH, f'/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/div[4]/div/div/div/table/tbody/tr[{y}]/td[1]'))).text
                print(project)
                projects.append(project)

                transaction = wait.until(EC.presence_of_element_located((By.XPATH, f'/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/div[4]/div/div/div/table/tbody/tr[{y}]/td[2]'))).text
                print(transaction)
                transactions.append(transaction)
                
                names2.append(name)

                y += 1

            except:
                project = 'No'
                y += 1

            b += 100
            driver.execute_script(f"window.scrollTo({str(a)},{str(b)})")

            if project == 'No':
                names2.append(name)
                projects.append('N/A')
                transactions.append('N/A')
                projectsTrue = False

        driver.find_element_by_tag_name('body').send_keys(Keys.CONTROL + Keys.HOME)

        print('------------------------------------------------------------')

        rankings_tab = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/nav/ul/li[5]/a')))
        driver.execute_script(f"window.scrollTo({str(0)},{str(0)})")
        rankings_tab.click()

        # rankings
        ranking = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[7]/div[3]/div/div[3]/div[1]/div/div/div[2]/div/div[5]/div[2]/div[2]/table/tbody/tr/td[2]/span'))).text
        print(ranking)
        names3.append(name)
        rankings.append(ranking)

        print('------------------------------------------------------------')

        driver.quit()
    
```
